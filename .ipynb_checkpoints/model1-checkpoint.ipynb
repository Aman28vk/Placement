```python
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv('placement data set.csv')
df
```

```python
# remove sl_no and salary column as we domnt need them as of now
df=df.drop(['sl_no', 'ssc_b','hsc_b', 'salary'], axis=1)
#or we can use "inplace=True" for saving the table in existing dataframe
```

```python
df
```

```python
df.isnull().sum()
```

```python
#check the data type of all column before using lanle encoder by 
df.dtypes
```

```python
df['gender']=df['gender'].astype('category')
df['degree_t']=df['degree_t'].astype('category')
df['specialisation']=df['specialisation'].astype('category')
df['workex']=df['workex'].astype('category')
df['status']=df['status'].astype('category')
df['hsc_s']=df['hsc_s'].astype('category')
df.dtypes
```

```python
#now using lable encoder for converrting the values into binary
df['gender']=df['gender'].cat.codes
df['degree_t']=df['degree_t'].cat.codes
df['specialisation']=df['specialisation'].cat.codes
df['workex']=df['workex'].cat.codes
df['status']=df['status'].cat.codes
df['hsc_s']=df['hsc_s'].cat.codes
```

```python
df.head()
```

```python
df.describe()
```

```python
#checking for outlier
columns = df.columns
fig, axes = plt.subplots(nrows=len(columns), figsize=(8, 6 * len(columns)))
for i, column in enumerate(columns):
    axes[i].boxplot(df[column])
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Value')
fig.tight_layout()
plt.show()
```

```python
#removie outlier as we can se outlier only in hsc_p and degree_p column
df=df[~(df['hsc_p']>=90)]
df=df[~(df['hsc_p']<=45)]
df=df[~(df['degree_p']>=85)]
```

```python
# again we will checking for outlier
columns = df.columns
fig, axes = plt.subplots(nrows=len(columns), figsize=(8, 6 * len(columns)))
for i, column in enumerate(columns):
    axes[i].boxplot(df[column])
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Value')
fig.tight_layout()
plt.show()
```

```python
df_class = df.copy()
x = df_class.iloc[:, 0:-1].values
y = df_class.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.18, random_state=0)
```

```python
sc= StandardScaler()
x_train = sc.fit_transform(x_train) # for standardising the features
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
log.score(x_train,y_train)
```
